在Python中创建迭代器最方便的技术是使用生成器。

```python
def factors(n):
    results = []
    for k in range(1, n+1):
        if n % k == 0:
            results.append(k)
    return results
```

```python
def factors(n):
    for k in range(1, n+1):
        if n % k == 0:
            yield k
```

```python
def factors(n):    # generator that computes factors
    k = 1
    while k * k < n:    # while k < sqrt(n)
        if n % k == 0:
            yield k
            yield n // k
        k += 1
    if k * k == n:    # special cese if n is perfect square
        yield k
```

```python
def fibonacci():
    a = 0
    b = 1
    while True:    # keeping going...
        yield a    # report value, a, during this case
        future = a + b
        a = b      # this will be next value repirted
        b = future    # and subsequently this
```
